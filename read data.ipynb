```python
import wfdb
import matplotlib.pyplot as plt
import os
dataset_path = "mit-bih/"
```

```python
# Load ECG signal from record 100
record = wfdb.rdsamp("mit-bih/121")  # Change path based on your dataset location
signals, fields = record
# Extract lead I ECG signal
ecg_signal = signals[:, 0]

# Plot the first 2000 samples for better visibility
plt.figure(figsize=(10, 4))
plt.plot(ecg_signal[:3600], color='black', linewidth=1)
plt.title("ECG Signal - Record 100 (Zoomed In)")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.show()
```

```python
for record_num in range(100, 235):
    record_path = os.path.join(dataset_path, str(record_num))

    try:
        # Load ECG signal
        record = wfdb.rdsamp(record_path)
        signals, fields = record

        # Get Sampling Frequency (Fs)
        fs = fields["fs"]
        samples_10s = fs * 10  # Get first 10 seconds

        # Extract first 10 seconds of Lead I
        ecg_signal = signals[:samples_10s, 0]

        # Plot ECG
        plt.figure(figsize=(10, 4))
        plt.plot(ecg_signal, color='black', linewidth=1)
        plt.axis('off')
        # Save each plot as an image
        plt.savefig(f"ecg_record_{record_num}.png", dpi=300, bbox_inches='tight', pad_inches=0)
        plt.close()  # Close figure to free memory

        print(f" Processed Record {record_num}")

    
    except Exception as e:
            print(f" Error processing Record {record_num}: {e}")
```

```python
save_folder = "org"
for record_num in range(100, 235):
    record_path = os.path.join(dataset_path, str(record_num))

    try:
        # Load ECG signal
        record = wfdb.rdsamp(record_path)
        signals, fields = record

        # Get Sampling Frequency (Fs)
        fs = fields["fs"]
        total_samples = len(signals)

        # Loop through the signal in 10-second segments
        segment_num = 1
        for start_idx in range(0, total_samples, fs * 10):
            end_idx = start_idx + fs * 10  # Define the end index
            if end_idx > total_samples:
                break  # Stop if there aren't enough samples left for a full 10s segment

            # Extract 10-second segment of Lead I
            ecg_signal = signals[start_idx:end_idx, 0]

            # Plot ECG
            plt.figure(figsize=(10, 4))
            plt.plot(ecg_signal, color='black', linewidth=1)
            plt.axis('off')

            # Generate filename and save in "org" folder
            segment_time = segment_num * 10
            image_filename = f"org_ecg_{record_num}_{segment_time}.png"
            image_path = os.path.join(save_folder, image_filename)
            plt.savefig(image_path, dpi=300, bbox_inches='tight', pad_inches=0)
            plt.close()  # Close figure to free memory

            print(f" Saved: {image_path}")
            segment_num += 1

    except Exception as e:
        print(f" Error processing Record {record_num}: {e}")
```
